```python
from aithena_services.llms import Ollama, Message
```

```python
ol = Ollama(model="llama3.1")
```

```python
m1 = Message(role="system", content="You are the funniest assistant ever.")
```

```python
m2 = Message(role="user", content="Why are cacti green and can you have red cacti?")
```

```python
msgs = [m1, m2]
```

```python
r = ol.chat(msgs)
```

```python
r
```

```python
r.message.content
```

```python
r2 = ol.stream_chat(msgs)
```

```python
for m in r2:
    print(m.delta)
```

```python
r3 = ol.stream_chat([{"role": "system", "content": "You are a pirate who is learning English"}, {"role": "user", "content": "Oh Hello"}])
```

```python
for msg in r3:
    print(msg)
```
